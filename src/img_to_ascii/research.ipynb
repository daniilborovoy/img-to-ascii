```python
import cv2
import numpy as np

ASCII_CHARS = "@%#*+=-:. "

def resize_image(image, new_width=100):
    height, width = image.shape
    aspect_ratio = height / width
    new_height = int(aspect_ratio * new_width * 0.5)
    resized = cv2.resize(image, (new_width, new_height))
    return resized

def grayify(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

def image_to_ascii(image, width=100):
    gray_image = grayify(image)

    resized_gray = resize_image(gray_image, new_width=width)

    normalized = resized_gray / 255.0

    ascii_img = ""
    for row in normalized:
        for pixel in row:
            ascii_img += ASCII_CHARS[int(pixel * (len(ASCII_CHARS) - 1))]
        ascii_img += "\n"
    
    return ascii_img

path = "Phillips-Neon-Genesis-Evangelion.webp"
image = cv2.imread(path)

ascii_art = image_to_ascii(image, width=120)
print(ascii_art)
```

# Dithering

```python
import cv2
import numpy as np

ASCII_CHARS = "@%#*+=-:. '  "


def grayify(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

def resize_image(image, new_width=100):
    height, width = image.shape[:2]  # works for grayscale or color
    aspect_ratio = height / width
    new_height = int(aspect_ratio * new_width * 0.5)
    return cv2.resize(image, (new_width, new_height))

def apply_dithering(gray_image):
    h, w = gray_image.shape
    output = gray_image.copy().astype(np.float32)
    num_chars = len(ASCII_CHARS)

    for y in range(h):
        for x in range(w):
            old_pixel = output[y, x]
            new_pixel_val = int((old_pixel / 255) * (num_chars - 1))
            new_pixel = new_pixel_val * (255 / (num_chars - 1))
            output[y, x] = new_pixel
            quant_error = old_pixel - new_pixel

            if x + 1 < w:
                output[y, x + 1] += quant_error * 7 / 16
            if y + 1 < h:
                if x > 0:
                    output[y + 1, x - 1] += quant_error * 3 / 16
                output[y + 1, x]     += quant_error * 5 / 16
                if x + 1 < w:
                    output[y + 1, x + 1] += quant_error * 1 / 16

    return np.clip(output, 0, 255).astype(np.uint8)

def image_to_ascii(image, width=100):
    gray = grayify(image)
    resized = resize_image(gray, new_width=width)
    dithered = apply_dithering(resized)

    ascii_img = ""
    for row in dithered:
        for pixel in row:
            ascii_img += ASCII_CHARS[int(pixel / 255 * (len(ASCII_CHARS) - 1))]
        ascii_img += "\n"
    return ascii_img

# Load image
path = "Phillips-Neon-Genesis-Evangelion.webp"
image = cv2.imread(path)

if image is None:
    raise ValueError(f"Image not found at path: {path}")

ascii_art = image_to_ascii(image, width=120)
print(ascii_art)
```
